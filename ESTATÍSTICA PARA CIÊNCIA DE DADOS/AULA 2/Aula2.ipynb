```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Eixo x entre 12 e 20 de .001 em .001.
x_axis = np.arange(12, 20, 0.001)
# Média = 15.75, desvio Padrão = 1.28
plt.plot(x_axis, norm.pdf(x_axis,15.75,1.28))
plt.xlabel("Hemoglobina (g/100ml)")
plt.ylabel("função densidade de probabilidade")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Eixo x entre 12 e 20 de .001 em .001.
x_axis = np.arange(12, 20, 0.001)
# Média = 15.75, desvio Padrão = 1.28
plt.plot(x_axis, norm.cdf(x_axis,15.75,1.28))
plt.xlabel("Hemoglobina (g/100ml)")
plt.ylabel("distribuição acumulada")
plt.show()
```

```python
from scipy.stats import bernoulli
import matplotlib.pyplot as plt
import numpy as np

p=0.5
n=1000
A=[sum(bernoulli.rvs(p, size=i))/i for i in range(5,n+1)]
x=np.arange(5,n+1)
A = np.column_stack((x,A))
plt.plot(A[:,0], A[:,1])

plt.xlabel('nº de repetições')
plt.ylabel('frequência relativa de acerto')
```
