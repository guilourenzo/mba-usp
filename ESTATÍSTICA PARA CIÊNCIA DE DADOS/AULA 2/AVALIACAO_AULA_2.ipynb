```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate
```

```python
def f(x):
    y = 20*x**3*(1-x)

    return y

x = np.linspace(0, 1, 50)
y = f(x)

plt.plot(x, y)
plt.axvline(0.3, 0, 1, linestyle = 'dashed', color='r')
plt.axvline(0.7, 0, 1, linestyle = 'dashed', color='r')
plt.show()
```

```python
Y = lambda x: 20*x**3*(1-x)
i = scipy.integrate.quad(Y, (1/3), (2/3))

print(i)
print(round(i[0], 4))
```
