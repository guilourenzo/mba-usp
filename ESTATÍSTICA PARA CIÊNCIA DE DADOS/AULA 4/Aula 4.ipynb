```python
# Conhecendo a forma da função de verossimilhança no exemplo da amostra da distribuição Normal
# tomando os dados do ENEM como exemplo
import pandas as pd
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/Mariana/Dropbox (Pessoal)/1º sem 2020/MBA/Dados/Dados_ENEM.csv", encoding= 'unicode_escape')

x = df.Media # iguala x a variável "Media" 
print(np.mean(x),np.std(x)) # calcula média e desvio padrão da variável

# Compara a verossimilhança da amostra para diferentes valores dos parâmetros mu
def L_amostra(x, mu, sd):
    L = 0
    for i in x:
        L += norm.pdf(i, mu, sd) # cálculo da função de verossimilhança da amostra
   
#    print ("média = ", mu, "desvio padrão = ", sd,"L:", L)
    return L


mu = np.linspace(200, 800, 1000)
L1 = L_amostra(x,mu,np.std(x)) # para a amostra de valores em x e para o desvio padrão suposto conhecido e igual ao amostral, calcula a verossimilhança para diversos valores de média
L2 = L_amostra(x,mu,20) # para a amostra de valores em x e para o desvio padrão sendo igual ao amostral, calcula a verossimilhança para diversos valores de média

plt.plot(mu,L1, label = r'$sigma^2=38^2$')
plt.plot(mu,L2, label = r'$sigma^2=20^2$')
plt.xlabel(r'$\mu$')
plt.ylabel(r'$L(\mu,\sigma^2)$')
plt.title(r'Média do ENEM: $N(\mu,\sigma^2)$')
plt.legend()
plt.show()
```

```python
#Conhecendo a forma da função log-verossimilhança no exemplo da amostra da distribuição Normal
# tomando os dados do ENEM como exemplo
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def get_ll_normal(x, mu, sigma):
    return np.sum(np.log(stats.norm.pdf(x, mu, sigma)))

x = df.Media # iguala x a variável "Media" 
sigma = np.std(x) # # parâmetro fixo, igualado ao desvio padrão amostral
theta = np.linspace(200, 800, 1000) # parâmetro que vamos variar (mu)
f_theta1 = np.array([get_ll_normal(x, mu, sigma) for mu in theta])

i_best = np.argmax(f_theta1) # cálculo (numérico) do estimador de máxima verossimilhança (EMV) de mu
print('Melhor mu:', theta[i_best])

plt.plot(theta, f_theta1, label = r'$\sigma^2=38^2$')
plt.plot([theta[i_best]], [f_theta1[i_best]], 'r*') # coloca uma estrelinha no maior valor (EMV)
plt.legend()
```

```python
from math import exp
import matplotlib.pyplot as plt
import numpy as np

# Compara EQM de dois estimadores de Sigma2
def func(sigma2,n):   #esta função calcula o EQM do estimador S2 (a demonstração de que o EQM(S2) é este não foi detalhada)
  y = []
  for l in sigma2:
    y.append(2*l**2/(n-1))

  return y

def func2(sigma2,n):   # esta função calcula o EQM do estimador Sigma2_chapéu (a demonstração de que o EQM(Sigma2_chapéu) é este não foi detalhada)
  y = []
  for l in sigma2:
    y.append((2*n-1)*l**2/(n**2))

  return y

n = 10
sigma2 = np.arange(1, 100)
eqm = func(sigma2,n)
eqm2 = func2(sigma2,n)
plt.xlabel(r'$\sigma^2$')
plt.ylabel(r'$EQM$')

plt.plot(sigma2, eqm, label = r'$S^2$')
plt.plot(sigma2, eqm2, label = r'$\hat{\sigma}^2$')
plt.legend()
plt.title('n=10')
plt.show()
```
