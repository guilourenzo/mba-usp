```python
import pandas as pd
df = pd.read_csv("C:/Users/Mariana/Dropbox (Pessoal)/1º sem 2020/MBA/Dados/Dados_ENEM.csv", encoding= 'unicode_escape')

df.head(3)
```

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt


# Generate some data for this demonstration.
data1 = df.loc[df.Admin=='Privada', "Media"]
data2 = df.loc[df.Admin=='Estadual', "Media"]

# Fit a normal distribution to the data:
mu1, std1 = norm.fit(data1)
mu2, std2 = norm.fit(data2)

# Plot the histogram.
#plt.subplots(figsize=(7,6), dpi=100)
plt.hist(data1, bins=25, density=True, color="dodgerblue", label="Privada; média = %.0f,  dp = %.0f" % (mu1, std1))
plt.hist(data2, bins=25, density=True, color="steelblue", label="Estadual; média = %.0f,  dp = %.0f" % (mu2, std2))

# Plot the PDF.
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p1 = norm.pdf(x, mu1, std1)
p2 = norm.pdf(x, mu2, std2)
plt.plot(x, p1, 'k', linewidth=2)
plt.plot(x, p2, 'k', linewidth=2)

plt.legend()

plt.show()
```

```python
from scipy.stats import binom   
n = 4
p = 0.5 #probabilidade de um experimento resultar em sucesso
x = range(0,n+1) # valores de x =0,1,2,3 ou 4
p_x = binom.pmf(x,n,p) #distribuição dos resultados
p_x
```

```python
import matplotlib.pyplot as plt
plt.xlabel('nº de acertos')
plt.ylabel('probabilidade')
for a,b in zip(x, p_x):
    plt.text(a-0.25, b-0.02, str(round(b,4)), color='white')
plt.title('Acertar ao acaso a ordem de preparo de 4 xícaras de chá')
plt.bar(x,p_x,color = (0.3,0.1,0.4,0.6))
```

```python
import scipy.stats as ss

def plot_exponential(x_range, mu=0, sigma=1, cdf=False, **kwargs):
    x = x_range
    if cdf:
        y = ss.expon.cdf(x, mu, sigma)
    else:
        y = ss.expon.pdf(x, mu, sigma)
    plt.plot(x, y, **kwargs)
x = np.linspace(0, 5, 5000)

plot_exponential(x, 0, 2, color='red', lw=2, ls='-', alpha=0.5, label=r'$\lambda$=0,5')
plot_exponential(x, 0, 1, color='purple', lw=2, ls='-', alpha=0.5, label=r'$\lambda$=1')
plot_exponential(x, 0, .5, color='blue', lw=2, ls='-', alpha=0.5, label=r'$\lambda$=1,5')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend();
```
