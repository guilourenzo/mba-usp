```python
import cv2
print('Versão da OpenCV: ', cv2.__version__, end='\n\n')
from matplotlib.pyplot import imshow
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

## Imagem original

```python
imagem = cv2.imread('wood.jpg')
img = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB) # equivalente a img = imagem[:,:,::-1] # inverte de BGR para RGB
#img=img[:,:,0] # pega apenas o primeiro canal da imagem
imshow(img)
#imshow(np.asarray(imagem[:, :, ::-1])) # altera a ultima dimensão para trocar azul por vermelho -> BRG to RGB
```

## Encontrando defeitos na madeira

Utilizacao de threshold e blur

```python
#img = img[::2,::2] # reduz imagem pela metade
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # equivalente a img = imagem[:,:,::-1] # inverte de BGR para RGB
w1 = cv2.blur(gray, (3, 3)) # aplica blur
_, w2 = cv2.threshold(w1, 120, 255, cv2.THRESH_BINARY_INV)
w3 = cv2.blur(w2, (21, 21)) # aplica blur
_, w4 = cv2.threshold(w3, 120, 255, cv2.THRESH_BINARY_INV)

plt.figure(figsize=(14, 9)) # LARGura e ALTura da imagem total em polegadas
plt.subplot2grid((2,2),(0,0)) #(linhas, colunas) do grid (l,c) do elemento
plt.title('Blur')
plt.imshow(w1, cmap='gray')

plt.subplot2grid((2,2),(0,1)) #(linhas, colunas) do grid (l,c) do elemento
plt.title('Blur')
plt.imshow(w2, cmap='gray')

plt.subplot2grid((2,2),(1,0)) #(linhas, colunas) do grid (l,c) do elemento
plt.title('Blur')
plt.imshow(w3, cmap='gray')

plt.subplot2grid((2,2),(1,1)) #(linhas, colunas) do grid (l,c) do elemento
plt.title('Threshold')
plt.imshow(w4, cmap='gray')
```
