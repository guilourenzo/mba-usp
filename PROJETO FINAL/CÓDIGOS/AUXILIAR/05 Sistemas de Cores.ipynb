```python
import cv2
print('Versão da OpenCV: ', cv2.__version__, end='\n\n')
from matplotlib.pyplot import imshow
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

## Imagem original

```python
imagem = cv2.imread('falcon.jpg')
img = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB) # equivalente a img = imagem[:,:,::-1] # inverte de BGR para RGB
imshow(img)
#imshow(np.asarray(imagem[:, :, ::-1])) # altera a ultima dimensão para trocar azul por vermelho -> BRG to RGB
```

## Sistema de Cores / Espaço de Cores

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(14, 8)) # LARGGura e ALTura da imagem total em polegadas #fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
plt.subplot2grid((2,2),(0,0)) #(linhas, colunas) do grid (l,c) do elemento
plt.title('Gray')
plt.imshow(gray, cmap='gray')

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.subplot2grid((2,2),(0,1)) #(linhas, colunas) do grid (l,c) do elemento
plt.title('HSV')
plt.imshow(hsv)

lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
plt.subplot2grid((2,2),(1,0)) #(linhas, colunas) do grid (l,c) do elemento
plt.title('L*a*b')
plt.imshow(lab)

YCrCb = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)
plt.subplot2grid((2,2),(1,1)) #(linhas, colunas) do grid (l,c) do elemento
plt.title('YCrCb')
plt.imshow(YCrCb)
```

## 5.1 Canais da imagem colorida

```python
(vermelho, verde, azul) = cv2.split(img)
plt.figure(figsize=(14, 8)) # LARGGura e ALTura da imagem total em polegadas #fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
plt.subplot2grid((2,2),(0,0)) #(linhas, colunas) do grid (l,c) do elemento
plt.title('Original')
plt.imshow(img)

plt.subplot2grid((2,2),(0,1)) #(linhas, colunas) do grid (l,c) do elemento
plt.title('Vermelho')
plt.imshow(vermelho, cmap="gray")

plt.subplot2grid((2,2),(1,0)) #(linhas, colunas) do grid (l,c) do elemento
plt.title('Verde')
plt.imshow(verde, cmap="gray")

plt.subplot2grid((2,2),(1,1)) #(linhas, colunas) do grid (l,c) do elemento
plt.title('azul')
plt.imshow(azul, cmap="gray")
```

## Exibindo os canais em formato colorido

```python
(canalVermelho, canalVerde, canalAzul) = cv2.split(img)
zeros = np.zeros(img.shape[:2], dtype = "uint8")
#zeros[:,:]=255 try it!!! Ao invés de zeros coloque na matriz valores 255

plt.figure(figsize=(14, 8)) # LARGGura e ALTura da imagem total em polegadas #fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
plt.subplot2grid((2,2),(0,0)) #(linhas, colunas) do grid (l,c) do elemento
plt.title('Original')
plt.imshow(img)

plt.subplot2grid((2,2),(0,1)) #(linhas, colunas) do grid (l,c) do elemento
plt.title('Vermelho')
plt.imshow(cv2.merge([canalVermelho, zeros, zeros]))

plt.subplot2grid((2,2),(1,0)) #(linhas, colunas) do grid (l,c) do elemento
plt.title('Verde')
plt.imshow(cv2.merge([zeros, canalVerde, zeros]))

plt.subplot2grid((2,2),(1,1)) #(linhas, colunas) do grid (l,c) do elemento
plt.title('Azul')
plt.imshow(cv2.merge([zeros, zeros, canalAzul]))
```
