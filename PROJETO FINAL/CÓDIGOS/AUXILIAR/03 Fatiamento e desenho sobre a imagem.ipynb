```python
import cv2
print('Versão da OpenCV: ', cv2.__version__, end='\n\n')
from matplotlib.pyplot import imshow
import numpy as np
%matplotlib inline
```

```python
imagem = cv2.imread('falcon.jpg')
(b, g, r) = imagem[234, 304] #veja que a ordem BGR e não RGB
print('O pixel (0, 0) tem as seguintes cores:')
print('Vermelho:', r, 'Verde:', g, 'Azul:', b)
print('Shape da imagem',imagem.shape) 
```

```python
# MANIPULANDO PIXELS
for y in range(100, imagem.shape[0], 2):
  for x in range(50, imagem.shape[1], 2):
    imagem[y, x] = (imagem[y, x, 0], 255, 255)
    #imagem[y, x] = (x, y, x)
    #imagem[y, x] = (0,(x*y)%256,(x*y)%256)
    #imagem[y:y+5, x: x+5] = (0,255,255)
imshow(np.asarray(imagem[:, :, ::-1]))
```

```python
for y in range(0, imagem.shape[0], 2): #percorre as linhas
  for x in range(0, imagem.shape[1], 2): #percorre as colunas
    imagem[y, x] = (0,   0,   (x+y)%256)
imshow(np.asarray(imagem[:, :, ::-1]))
```

```python
#SLICING...
imagem[30:50, :] = (255, 0, 0) #azul
imagem[30:50, 50:100] = (255, 255, 0) #ciano
imagem[80:100, 50:100] = (255, 0, 255) # Rosa
imagem[200:230, 120:150] = (0, 255, 255) # amarelo
imshow(np.asarray(imagem[:, :, ::-1]))
```

```python
imagem = cv2.imread('falcon.jpg')
(X, Y) = (imagem.shape[1] // 2, imagem.shape[0] // 2)
for raio in range(0, 175, 5):
    cv2.circle(imagem, (X, Y), raio, (0,255,0))

fonte = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(imagem,'Polo Inovale',(25,300), fonte, 2, 
            (255,255,255),30,cv2.LINE_AA)
cv2.putText(imagem,'Polo Inovale',(25,300), fonte, 2, 
            (0,0,255),5,cv2.LINE_AA)
imshow(np.asarray(imagem[:, :, ::-1]))
cv2.imwrite('aiaiai.png', imagem)
```
