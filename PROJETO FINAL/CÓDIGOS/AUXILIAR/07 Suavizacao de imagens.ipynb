```python
import cv2
print('Versão da OpenCV: ', cv2.__version__, end='\n\n')
from matplotlib.pyplot import imshow
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

## Imagem original

```python
imagem = cv2.imread('falcon.jpg')
img = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB) # equivalente a img = imagem[:,:,::-1] # inverte de BGR para RGB
imshow(img)
#imshow(np.asarray(imagem[:, :, ::-1])) # altera a ultima dimensão para trocar azul por vermelho -> BRG to RGB
```

## 7.1 Suavização por cálculo da média (Blur)

```python
plt.figure()
plt.title('Original')
plt.imshow(img)

plt.figure(figsize=(14, 14)) # LARGura e ALTura da imagem total em polegadas
box = 3
for linha in range(3):
    for coluna in range(2):
        plt.subplot2grid((3,2),(linha,coluna)) #(linhas, colunas) do grid (l,c) do elemento
        plt.title('Blur '+str(box)+'x'+str(box))
        plt.imshow(cv2.blur(img, (box,box)))
        box+=2
```

## 7.2 Suavização pela Gaussiana

```python
plt.figure()
plt.title('Original')
plt.imshow(img)

plt.figure(figsize=(14, 14)) # LARGura e ALTura da imagem total em polegadas
box = 3
for linha in range(3):
    for coluna in range(2):
        plt.subplot2grid((3,2),(linha,coluna)) #(linhas, colunas) do grid (l,c) do elemento
        plt.title('Gaussian Blur '+str(box)+'x'+str(box))
        plt.imshow(cv2.GaussianBlur(img, (box,box), 0))
        box+=2
```

## 7.3 Suavização pela mediana

```python
plt.figure()
plt.title('Original')
plt.imshow(img)

plt.figure(figsize=(14, 14)) # LARGura e ALTura da imagem total em polegadas
box = 3
for linha in range(3):
    for coluna in range(2):
        plt.subplot2grid((3,2),(linha,coluna)) #(linhas, colunas) do grid (l,c) do elemento
        plt.title('Median Blur '+str(box)+'x'+str(box))
        plt.imshow(cv2.medianBlur(img, box))
        box+=2
```

## 7.4 Suavização com filtro bilateral

```python
plt.figure()
plt.title('Original')
plt.imshow(img)

plt.figure(figsize=(14, 14)) # LARGura e ALTura da imagem total em polegadas
box = 3
for linha in range(3):
    for coluna in range(2):
        plt.subplot2grid((3,2),(linha,coluna)) #(linhas, colunas) do grid (l,c) do elemento
        plt.title('Bilateral Filter Blur '+str(box)+'x'+str(box))
        plt.imshow(cv2.bilateralFilter(img, box, box*7, box*7))
        box+=2
```
